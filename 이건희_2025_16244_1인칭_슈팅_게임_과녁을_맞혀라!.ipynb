# 프로그램 기획서: 1인칭 슈팅 게임: 과녁을 맞혀라!

## 1.프로젝트 요약(공백포함 500글자 이내)




- 목적/내용/결론

어릴 때부터 게임을 좋아했고, 특히 FPS 게임을 즐겨해 이번 프로젝트를 통해 투사체를 던져 과녁을 히는 1인칭 슈팅 게임을 제작해보았습니다. 레이캐스팅 기법을 통해 2차원 맵을 3D처럼 보이도록 구현하였으며, 맵에 주어진 과녁을 모두 맞출 경우 클리어 화면이 뜨며 자동으로 게임이 종료됩니다. 플레이 중 전후좌우 이동 및 점프가 가능하고, 마우스 휠을 통해 시점 전환이 가능하며, 마우스를 움직이면 시야를 조정할 수 있습니다. 마우스 좌클릭 시 투사체가 발사되도록 하였고, 마우스 우클릭 시 시야를 확대할 수 있습니다. 화면의 정중앙에는 조준점을 배치하여 조준이 편리하도록 하였으며, 좌측 상단에는 미니맵을 만들어 현재 플레이어의 위치와 과녁의 위치를 확인할 수 있도록 하였습니다. 배경음악은 GarageBand로 간단하게 작곡하여 추가하였고, 수업 시간에 활용한 효과음으로 사운드 효과를 추가하여 게임의 몰입도를 높였습니다. 이 게임은 맵을 이동하며 모든 과녁을 맞히면 승리합니다.

## 2.기능
- 기획하고 있는 프로그램의 세부 기능을 번호를 매겨 작성해주세요.


1. 이동기능
- w,a,s,d 키로 전후좌우 이동이 가능하며, 스페이스 바로 점프가 가능하다.
2. 시야 조정
- 마우스를 이동하여 상하좌우로 시야를 조정할 수 있다.
3. 투사체 발사
- 마우스 좌클릭으로 투사체를 발사할 수 있으며, 사운드 효과가 동반된다. 투사체는 직진하여 벽에 부딪히면 소멸한다.
4. zoom 기능
- 마우스 우클릭으로 확대가 가능하다.
5. 1인칭-3인칭 시점 전환 기능
- 마우스 휠을 내리면 3인칭으로 시점이 전환되며, 휠을 올리면 1인칭으로 시점이 전환된다.
6. 배경음악 정지/재생 기능
- p 키를 누르면 배경음악이 꺼지고, 다시 p키를 누르면 배경음악이 재생된다.
7. 게임 (수동)종료 기능
- esc 키를 누르면 게임이 종료된다.
8. 미니맵 기능
- 미니맵을 통해 플레이어의 위치와 과녁의 위치를 확인할 수 있다.
9. 조준점 기능
- 화면 정중앙에 조준점을 배치해 조준의 편리성을 높인다.
10. 과녁 및 점수 기능
- 투사체로 과녁을 맞히면 과녁이 소멸되고, 점수가 증가한다.
11. 자동 종료 기능
- 플레이어가 모든 과녁을 맞췄다면, 게임이 자동으로 종료된다.

## 2.실행방법

- 기획서에 작성한 기능을 테스트할 수 있도록 모든 기능에 대한 테스트케이스를 적어주세요.


게임을 실행하면 플레이어는 w,a,s,d, 스페이스바 키를 통해 이동할 수 있다. 마우스를 움직이면 시야를 조정할 수 있고, 마우스 휠을 내리면 3인칭으로 시점이 전환되며, 휠을 올리면 1인칭 시점으로 전환된다. 마우스 좌클릭을 하면 투사체가 발사되며, 사운드 효과가 발생한다. 마우스 우클릭을 하면 특정 지점을 확대(zoom) 할 수 있다. 배경음악은 p를 누르면 정지되며, 다시 p를 누르면 재생된다. esc키를 누르면 게임이 즉시 종료된다. 맵은 벽으로 둘러싸여 있고 어떤 벽에는 과녁이 붙어있다. 과녁을 투사체로 맞히면 과녁은 소멸하며, 점수가 올라간다. 각 과녁은 점수가 다르다. (빨간색은 10점, 초록색은 20점, 파란색은 30점) 맵에 있는 모든 과녁을 소멸시켰다면 게임 클리어 화면이 뜨며 점수가 뜬다. 2초가 지나면 게임이 종료된다.

## 3.AI 질문 리스트
- 생성형 AI의 답변을 활용한 경우에는 질문과 답변을 어떻게 활용했는지 작성해주세요.

먼저 1인칭 슈팅 게임을 만들기 위해서는 3D를 구현해야겠다고 생각했습니다. 생성형 AI(GPT)에게 python으로 3D를 구현할 수 있는 방법에 대해 물었고, 레이캐스팅 방식을 통해 2D맵을 3D처럼 보이도록 할 수 있다는 답변을 얻었습니다.

그 후 플레이어가 전후좌우로 이동할 수 있는 3D맵을 레이캐스팅을 통해 구현하는 코드를 질문하여 게임의 전체적 틀을 잡았고, 기능을 하나씩 추가해 나갔습니다.

먼저 점프할 수 있는 기능을 추가할 수 있도록 하였습니다. 그런데 점프를 하면 벽도 같이 상승해서, 그 원인이 되는 코드가 무엇인지 질문하고, 점프를 할 때는 주변은 하강하는 것처럼 보이도록 수정하였습니다.

또한 시야가 처음에는 좌우 이동으로 한정되어 있어 관련 코드를 어떻게 수정하면 좋을지 질문하여 상하좌우 이동으로 변경하였고, 조준점을 추가해 시야의 중심을 잡아주도록 하였습니다.

게임을 위해 과녁을 추가하였고, 공중에 떠있는 것이 아닌 벽에 붙어있을 수 있도록 조정하기 위해 어떻게 수정해야 하는지 답변을 얻었습니다.

또한 투사체가 과녁에 맞아도 사라지지 않거나, 맞지 않았는데 과녁이 사라지는 문제를 해결하기 위해 과녁이 벽보다 조금 더 앞에 존재하도록 위치를 조정하고, 충돌로 판정되는 과녁의 반지름을 조정하면 된다는 답변을 얻었습니다.

zoom 기능을 추가하였을 때는 벽이 가까이 보이는 것이 아니라 좌우로 늘어나는 것처럼 보여 이를 조정하기 위해 FOV 보간과 투영면 계산을 적용해야 한다는 답변을 받았습니다. 게임을 클리어 하면 자동 종료하는 기능을 구현하는 법에 대해 질문하고, 그에 대한 답변을 얻었습니다.
